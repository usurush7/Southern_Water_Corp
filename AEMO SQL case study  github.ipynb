<b> i) Write a SQL Statement showing the `Total` of all Outage Types (Forced, Consequential, Scheduled, Opportunistic) where the `Status = Approved`, that occurred for both 2016 and 2017, grouped by `Year` and `Month`.  per month (i.e. 1 ‚Äì 12). Order by `Year`, `Month`, `Total_Number_Outages` in Descending Order.


ii) Building on the query you write in i), group the results by `Outage Type`, `Year` and `Month`. This is so you can identify whether there is any outage type specifically increasing on a monthly basis when comparing 2016 to 2017. </b>



```python
%%sql
WITH OutageSummary AS (
    SELECT 
        Year,
        Month,
        COUNT(*) AS Total_Number_Outages
    FROM 
        AEMR_Outage_Table
    WHERE 
        Status = 'Approved' AND Year IN (2016, 2017)
    GROUP BY 
        Year, Month
)
SELECT 
    Year,
    Month,
    SUM(Total_Number_Outages) AS Total_Outages
FROM 
    OutageSummary
GROUP BY 
    Year, Month
ORDER BY 
    Year, Month, Total_Outages DESC;
```

<h3 style="color:#D4AF37"> Question Three </h3>

<b>Write a SQL statement that calculates 1) The `Total_Number_Outage_Events` and 2) The <b> `Average Duration`</b> in <u>DAYS</u> for each `Participant Code` and `Outage Type` over the 2016 and 2017 Period where the `Status = Approved`. 
Order by `Total_Number_Outage_Events` in Descending Order, `Reason` and `Year`.
    
Please note the average duration in days should be rounded to 2 decimal places for ease of comparison. When calculating the average duration, please note that you'll need to use the following fields:

`Start_Time` and `End_Time`. </b>




```python
%%sql
WITH OutageSummary AS (
    SELECT 
        Participant_Code,
        Outage_Reason,
        COUNT(*) AS Total_Number_Outage_Events,
        AVG(ROUND((JULIANDAY(End_Time) - JULIANDAY(Start_Time)), 2)) AS Average_Outage_Duration_In_Days
    FROM 
        AEMR_Outage_Table
    WHERE 
        Status = 'Approved' 
        AND Year IN (2016, 2017)
    GROUP BY 
        Participant_Code, Outage_Reason
)
SELECT 
    Participant_Code,
    Outage_Reason,
    Total_Number_Outage_Events,
    Average_Outage_Duration_In_Days
FROM 
    OutageSummary
ORDER BY 
    Total_Number_Outage_Events DESC,
    Outage_Reason,
    Participant_Code;
```

<h3 style="color:#D4AF37"> Question Four </h3>

<b> Using the above criteria for context, write a SQL Statement that <u> classifies each participant code as either `High Risk`, `Medium Risk` or `Low Risk` in a column called `Risk_Classification`</u> that is based off their Average Outage Duration Time. Please note that this is for all valid (i.e. `Where status = approved`) outage types (Forced, Consequential, Scheduled, Opportunistic) for <u>all</u> participant codes from 2016 to 2017. Order the results using `Average Duration Time In Days` in descending order. 



```python
%%sql
WITH OutageSummary AS (
    SELECT 
        Participant_Code,
        Outage_Reason,
        COUNT(*) AS Total_Number_Outage_Events,
        AVG(ROUND((JULIANDAY(End_Time) - JULIANDAY(Start_Time)), 2)) AS Average_Outage_Duration_In_Days
    FROM 
        AEMR_Outage_Table
    WHERE 
        Status = 'Approved' 
        AND Year IN (2016, 2017)
        AND Outage_Reason IN ('Forced', 'Consequential', 'Scheduled', 'Opportunistic')
    GROUP BY 
        Participant_Code, Outage_Reason
), RiskClassification AS (
    SELECT 
        Participant_Code,
        AVG(Average_Outage_Duration_In_Days) AS Average_Outage_Duration_In_Days
    FROM 
        OutageSummary
    GROUP BY 
        Participant_Code
)
SELECT 
    os.Participant_Code,
    os.Outage_Reason,
    os.Total_Number_Outage_Events,
    rc.Average_Outage_Duration_In_Days,
    CASE 
        WHEN rc.Average_Outage_Duration_In_Days > 1 THEN 'High Risk'
        WHEN rc.Average_Outage_Duration_In_Days >= 0.5 AND rc.Average_Outage_Duration_In_Days <= 1 THEN 'Medium Risk'
        ELSE 'Low Risk'
    END AS Risk_Classification
FROM 
    OutageSummary os
JOIN 
    RiskClassification rc ON os.Participant_Code = rc.Participant_Code
ORDER BY 
    rc.Average_Outage_Duration_In_Days DESC;


```

<h3 style="color:#D4AF37"> Question Five </h3>

<b> Just as you did in Question Four, Using the above criteria for context, write a SQL Statement that <u> classifies each participant code as either `High Risk`, `Medium Risk` or `Low Risk` in a column called `Risk_Classification`</u> using the new classification criteria. Order the results using `Average Duration Time In Days` in descending order. 




```python
%%sql
WITH OutageSummary AS (
    SELECT 
        Participant_Code,
        SUM(CASE WHEN Outage_Reason = 'Forced' THEN 1 ELSE 0 END) AS Forced_Outages,
        COUNT(*) AS Total_Number_Outage_Events,
        AVG(ROUND((JULIANDAY(End_Time) - JULIANDAY(Start_Time)), 2)) AS Average_Outage_Duration_In_Days
    FROM 
        AEMR_Outage_Table
    WHERE 
        Status = 'Approved' 
        AND Year IN (2016, 2017)
    GROUP BY 
        Participant_Code
)
SELECT 
    os.Participant_Code,
    os.Total_Number_Outage_Events,
    os.Average_Outage_Duration_In_Days,
    CASE 
        WHEN os.Forced_Outages > 20 THEN 'High Risk'
        WHEN (os.Forced_Outages BETWEEN 10 AND 20) THEN 'Medium Risk'
        ELSE 'Low Risk'
    END AS Risk_Classification
FROM 
    OutageSummary os
ORDER BY 
    os.Average_Outage_Duration_In_Days DESC;



```

<h3 style="color:#D4AF37"> Question Six </h3>

<b> Write a SQL Statement to calculate the proportion of Forced Outages that have occurred over the 2016 - 2017 Period.
Do we observe any particular increases regarding any Outage Types over this period? </b>

```python
%%sql
WITH ForcedOutageSummary AS (
    SELECT 
        COUNT(*) AS Total_Number_Forced_Outage_Events
    FROM 
        AEMR_Outage_Table
    WHERE 
        Status = 'Approved'
        AND Year IN (2016, 2017)
        AND Outage_Reason = 'Forced'
),
TotalOutageSummary AS (
    SELECT 
        COUNT(*) AS Total_Number_Outages
    FROM 
        AEMR_Outage_Table
    WHERE 
        Status = 'Approved'
        AND Year IN (2016, 2017)
)
SELECT 
    t.Total_Number_Outages,
    f.Total_Number_Forced_Outage_Events,
    ROUND((f.Total_Number_Forced_Outage_Events * 100.0 / t.Total_Number_Outages), 2) AS Pct_Outage_Forced
FROM 
    TotalOutageSummary t,
    ForcedOutageSummary f;
```

<h3 style="color:#D4AF37"> Question Seven </h3>

Write a SQL Statement to calculate the `Total Number of Outages`, `Total Duration In Days`and `Total Energy Lost` of all valid `Outages` for each `participant code` and `facility_code`, sorted by `Total Energy Lost` in descending order and Ordered by the YEAR Category. 
    

```python
%%sql
WITH OutageSummary AS (
    SELECT 
        Participant_Code,
        Facility_Code,
        YEAR,
        Outage_Reason,
        COUNT(*) AS Total_Number_Outages,
        SUM(ROUND((JULIANDAY(End_Time) - JULIANDAY(Start_Time)), 2)) AS Total_Duration_In_Days,
        SUM(Energy_Lost_MW) AS Energy_Lost_MW
    FROM 
        AEMR_Outage_Table
    WHERE 
        Status = 'Approved'
    GROUP BY 
        Participant_Code, Facility_Code, YEAR, Outage_Reason
)
SELECT 
    Total_Number_Outages,
    Total_Duration_In_Days,
    Energy_Lost_MW,
    Outage_Reason,
    Participant_Code,
    Facility_Code,
    Year
FROM 
    OutageSummary
ORDER BY 
    Energy_Lost_MW DESC,
    Year;

```

<h3 style="color:#D4AF37"> Question Eight </h3>

<b> Write a SQL Statement to calculate the `Average Duration In Days`and `Average Energy Lost` of all valid `FORCED OUTAGES` for each `participant code` and `facility_code` sorted by `Average Energy Lost` in descending order and Ordered by the YEAR Category. 


```python
%%sql
WITH ForcedOutageSummary AS (
    SELECT 
        Participant_Code,
        Facility_Code,
        YEAR,
        Outage_Reason,
        AVG(ROUND((JULIANDAY(End_Time) - JULIANDAY(Start_Time)), 2)) AS Avg_Duration_In_Days,
        AVG(Energy_Lost_MW) AS Avg_Energy_Lost
    FROM 
        AEMR_Outage_Table
    WHERE 
        Status = 'Approved'
        AND Outage_Reason = 'Forced'
    GROUP BY 
        Participant_Code, Facility_Code, YEAR, Outage_Reason
)
SELECT 
    Avg_Duration_In_Days,
    Avg_Energy_Lost,
    Outage_Reason,
    Participant_Code,
    Facility_Code,
    Year
FROM 
    ForcedOutageSummary
ORDER BY 
    Avg_Energy_Lost DESC,
    Year;
```

<h3 style="color:#D4AF37"> Question Nine </h3>

<b> Write a SQL Statement to calculate the `Average Energy Lost` and `Total Energy Lost` for each `Facility Code` and `Participant Code` across both the 2016 and 2017 periods when the `Outage_Reason` is set to Forced. Upon completion of this, calculate the <u> percentage </u> of energy lost due to forced outages for each `Facility_Code`. Please ORDER the query by `Total Energy Lost` from 2016 to 2017.
    
From your analysis, which participants have contributed the most to the Energy Lost due to Forced Outages?


```python
%%sql
WITH ForcedOutageSummary AS (
    SELECT 
        Participant_Code,
        Facility_Code,
        SUM(Energy_Lost_MW) AS Total_Energy_Lost,
        AVG(Energy_Lost_MW) AS Avg_Energy_Lost
    FROM 
        AEMR_Outage_Table
    WHERE 
        Status = 'Approved'
        AND Outage_Reason = 'Forced'
        AND Year IN (2016, 2017)
    GROUP BY 
        Participant_Code, Facility_Code
),
TotalEnergyLost AS (
    SELECT 
        Facility_Code,
        SUM(Energy_Lost_MW) AS Total_Energy_Lost_2016_2017
    FROM 
        AEMR_Outage_Table
    WHERE 
        Status = 'Approved'
        AND Outage_Reason = 'Forced'
        AND Year IN (2016, 2017)
    GROUP BY 
        Facility_Code
)
SELECT 
    f.Avg_Energy_Lost AS Avg_Energy_Lost,
    f.Total_Energy_Lost AS Total_Energy_Lost,
    ROUND((f.Total_Energy_Lost * 100.0 / t.Total_Energy_Lost_2016_2017), 2) AS Pct_Energy_Loss,
    'Forced' AS Outage_Reason,
    f.Participant_Code AS Participant_Code,
    f.Facility_Code AS Facility_Code,
    NULL AS Year
FROM 
    ForcedOutageSummary f
JOIN 
    TotalEnergyLost t ON f.Facility_Code = t.Facility_Code
ORDER BY 
    t.Total_Energy_Lost_2016_2017 DESC;

     

```

<h3 style="color:#D4AF37"> Question Ten </h3>

<b>Having identified the top 3 participants by Total Energy Loss being `GW`, `MELK` and `Auricon`; Write a SQL Statement calculating the `Total_Energy_Lost` each of these three `Participant_Codes` and the `Facility_Code`. Additionally, identify the `Description_Of_Outage` associated with the highest `Total_Energy_Lost` for each of the `Participant_Codes` and `Facility_Code` for each of the three participants. 
    
<u>Lastly, calculate the percentage of Energy Loss, attributed to these reasons!</u> </b>



```python
%%sql
WITH TopParticipants AS (
    SELECT 
        Participant_Code,
        Facility_Code,
        Description_Of_Outage,
        SUM(Energy_Lost_MW) AS Total_Energy_Lost,
        ROW_NUMBER() OVER(PARTITION BY Participant_Code ORDER BY SUM(Energy_Lost_MW) DESC) AS rank,
        SUM(SUM(Energy_Lost_MW)) OVER(PARTITION BY Participant_Code) AS Total_Energy_Lost_Participant
    FROM 
        AEMR_Outage_Table
    WHERE 
        Status = 'Approved'
        AND Participant_Code IN ('GW', 'MELK', 'Auricon')
        AND Year IN (2016, 2017)
    GROUP BY 
        Participant_Code, Facility_Code, Description_Of_Outage
)
SELECT 
    Participant_Code,
    Facility_Code,
    Description_Of_Outage,
    Total_Energy_Lost,
    ROUND((Total_Energy_Lost * 100.0 / Total_Energy_Lost_Participant), 2) AS Pct_Energy_Loss,
    rank
FROM 
    TopParticipants
ORDER BY 
    Participant_Code, rank;

```
